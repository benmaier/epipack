```python
import numpy as np
import epipack as epk
import matplotlib.pyplot as pl
%matplotlib notebook
```

```python
latent_period = 7
infectious_period = 8

omega = 1/latent_period
rho = 1/infectious_period

E0 = 1e-4

R0 = 2

alpha = rho * R0

pdf = lambda t: omega*rho/(omega-rho) * (np.exp(-t*rho) - np.exp(-t*omega))

T = np.linspace(0,60,30)
pl.plot(T, pdf(T))
```

```python
infcurve = 0.5 * (pdf(T)[1:] + pdf(T)[:-1])
```

```python
taus = np.diff(T)
norm = infcurve.dot(taus)
infcurve = infcurve/norm * R0
```

```python
pl.step(T[:-1],infcurve,where='post')
```
